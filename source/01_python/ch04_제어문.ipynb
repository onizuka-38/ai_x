# 1절. 조건문(if)
```
if 조건1:
    조건1이 참일 경우 수행할 명령어들
elif 조건2:
    조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들
else:
    모든 조건을 만족하지 않을 때 수행할 명령어들

```

````python
score = int(input('점수는 :'))
if score >= 90:
    print('참 잘했어요')
    print('다음에도 기대할게요')
elif score >= 60:
    print('분발하세요')
    print('뭐 그래도 고생하셨어요')
else:
    print('재시험 응시하셔야 합니다. \n화이팅')
````

````python
# 점수(0점~100점)를 입력받아 학점을 출력하시오(A, B, C, D, F, 유효하지 않은 점수)
score = int(input('점수는 :'))
if 90<=score<=100:      # if (score >=90) & (score<=100):
    grade = 'A'
elif 80<=score<90:
    grade = 'B'
elif 70<=score<80:
    grade = 'C'
elif 60<=score<70:
    grade = 'D'
elif 0<=score<60:
    grade = 'F'
else:
    grade = '유효하지 않은 점수'
print("입력하신 점수는 {}점이고, {}등급입니다.".format(score,grade))
````

# 2절. 반복문
## 2.1 for-in문
```
for 임시변수 in 나열가능한 자료:
    반복할명령어1
    반복할명령어2
else:
    나열가능한 자료가 다 소요하여 for블록을 다 수행한 후 최종 한번 수행할 명령어
    for문안에 break를 수행할 경우 else블록은 수행되지 않음
```

````python
for i in range(1, 10, 2):
    print(i, end='\t')
else:
    print('\nDone')
print('End')
````

````python
for i in range(1, 10, 2):
    if i ==5:
        break
    print(i, end='\t')
else:
    print('\nDone')
print('End')
````

````python
# 1~100까지 누적합을 출력하시오(for문)
tot = 0  # 누적변수
for i in range(1, 101):
    tot += i # tot = tot + i
print('1~100까지 합은 :', tot)
````

## 2.2 while문
```
while 조건:
    조건이 참일 경우 반복할 문장들
else 조건:
    조건이 거짓일 경우 한번 실행할 문장(else구문 선택사항)
```

````python
# 1~100까지 누적합 출력
tot = 0
i = 1
while i<=100:
    tot += i
    i += 1
print("1~100까지 누적합은", tot)
i
````

````python
# 1~100까지 짝수누적합 출력
tot = 0
i = 1
while i<=100:
    if i%2==0:
        tot += i
    i += 1
print("1~100까지 누적합은", tot)
i
````

````python
### ex1. 사용자에게 두 정수를 입력받아, 작은수부터 큰수까지 홀수의 합을 누적해서 출력

num1 = int(input('첫번째 숫자를 입력하세요: '))
num2 = int(input('두번째 숫자를 입력하세요: '))

start = min(num1, num2)
end = max(num1, num2)

tot = 0

for i in range(start, end + 1):
    if i%2:     # i % 2 == 1:
        tot += i

print('{}부터 {}까지 홀수들의 합은 {}입니다.'.format(start, end, tot))
````

````python
'''
10. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
(프로그램안에 임의의 숫자가 만들어지고 그 숫자를 맞추도록 합니다)
1단계 . 입력된 숫자가 틀리면 UP/DOWN 메시지를 출력하고 숫자를 맞추면 축하메세지와 함께 종료합니다(맞출 때까지 반복)
'''

import random
# (1) 1~100사이의 난수 발생
randomNum = random.randint(1,100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)

# (2) 수를 입력받아 틀리면 UP/DOWN 메시지를 출력, 맞추면 축하축하 메세지와 break하는 반복문
while True:
    inputNum = int(input("1~100사이의 수를 맞춰 보세요 :"))
    if (inputNum<1) | (inputNum>100):
        print('수의 범위를 벗어났습니다')
    elif randomNum < inputNum:
        print('DOWN ↓')
    elif randomNum > inputNum:
        print('UP ↑')
    else:
        print('축하축하 맞추셨습니다~')
        break
print('끝~')
````

# 4절. 중첩루프 탈출

````python
# 중첩루프 탈출 하는 방법 1 : 플래그 변수 이용
flag = False
for a in range(0, 3):         # a = 0, 1, 2
    for b in range(1, 3):     # b = 1, 2
        if a == b:
            flag = True
            break  # 반복문을 아예 빠져 나오기
        print(a,b)
    # break시 빠져나오는 부분
    if flag:
        break
print('DONE')
````

````python
# 중첩 루프 탈출 방법 2 : 예외 발생 => 예외처리
class BreakOutError(Exception):
    pass
try:
    for a in range(0,3): # a=0, 1, 2
        for b in range(1,3): # b=1, 2
            if a==b:
                raise BreakOutError() # 중첩 루프 빠져나가기
            print(a,b)
except BreakOutError:
    print('DONE')    
````

````python
'''
10. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
(프로그램안에 임의의 숫자가 만들어지고 그 숫자를 맞추도록 합니다)
1단계 . 입력된 숫자가 틀리면 UP/DOWN 메시지를 출력하고 숫자를 맞추면 축하메세지와 함께 종료합니다(맞출 때까지 반복)
'''

import random
# (1) 1~100사이의 난수 발생
randomNum = random.randint(1,100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)

min_n = 1
max_n = 100
count = 5

# (2) 수를 입력받아 틀리면 UP/DOWN 메시지를 출력, 맞추면 축하축하 메세지와 break하는 반복문
while count>0:
    print(f"{min_n}~{max_n} 사이의 수를 맞춰 보세요 (남은 기회는 {count}번)")
    inputNum = int(input("입력 : "))
    
    if inputNum < min_n or inputNum > max_n:
        print(f"수의 범위를 벗어났습니다! 수의 범위는 {min_n}~{max_n}입니다.")
        continue
        
    if inputNum > randomNum:
        print("Down ↓")
        max_n = inputNum -1
        count = count -1
        
    elif inputNum < randomNum:
        print("UP ↑")
        min_n = inputNum + 1
        count = count -1
        
    else:
        print('축하축하 맞추셨습니다~')
        break
print('끝~')

if count == 0:
    print()
    print("기회를 모두 사용했습니다. 실패하셨습니다.")
````
